```python
from datasets import load_dataset
import os
import soundfile as sf

output_dir = "/home/sidleal/sources/data/cml-tts"
mls = load_dataset("ylacombe/cml-tts", "portuguese", split="dev", streaming=False)
#print(next(iter(mls)))

fm = open(output_dir + "/metadata_dev.csv", "w", encoding="utf-8")
fm.write("path\tspeaker\tsample_rate\ttext\n")

batch_size = 1000
for i, batch in enumerate(mls.iter(batch_size=batch_size)):
    print(f"Processing batch {i+1}")
    #print(batch)
    for j in range(0, len(batch['text'])):
        #print(batch['text'][j])
        text = batch['text'][j]
        speaker_id = batch['speaker_id'][j]
        audio = batch['audio'][j]['array']
        path = batch['audio'][j]['path']
        sample_rate = batch['audio'][j]['sampling_rate']
        #print(f"Text: {text} Audio: {audio} sample_rate: {sample_rate} path: {path}")
        fm.write(f"{path}\t{speaker_id}\t{sample_rate}\t{text}\n")
        os.makedirs(f"{output_dir}/{speaker_id}", exist_ok=True )
        filename = os.path.join(f"{output_dir}/{speaker_id}", path)
        sf.write(filename, audio, sample_rate)
        print(f"Saved audio to: {filename}")
fm.close()
```
